```python
from roboflow import Roboflow
from ultralytics import YOLO
from PIL import Image
import matplotlib.pyplot as plt
```

```python
from ultralytics import YOLO
model_pretrained = YOLO("/content/yolo11n.pt") # Pretreained model contain clusters
```

```python
image_path = "/content/image.jpg"
image = Image.open(image_path)

plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
results = model_pretrained(image)

# Plot detection boundaires
detection_image = results[0].plot(show=False)  # Fence image detected
plt.imshow(detection_image)  # Show results with matplotlib
plt.axis('off')
plt.show()
```

```python
from ultralytics import YOLO
model = YOLO("/content/yolo11n.pt")
```

```python

results = model.train(
    data='/content/data.yaml',
    epochs=50,
    batch=16,
    name='yolo11n_training'
)
```

```python

metrics = model.val(data='/content/data.yaml',
                     name='yolo11n_val')
```

```python
from IPython.display import Image, display
# Show the training results (like loss, metrics)
display(Image(filename='/content/runs/detect/yolo11n_val/confusion_matrix_normalized.png'))
```
